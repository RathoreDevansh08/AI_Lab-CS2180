```python
import numpy as np
```

```python
file=open("/home/111701011/Downloads/lab3/speeches.txt","r")
```

```python
words=[]
sent=[]
data=file.readlines()
for line in data:
    words+=line.split()
```

```python
dicta={}

for i in range(len(words)):
    word=words[i]
    if(i==len(words)-1):
        break
    if word in dicta:
        dicta[word].append(words[i+1])
    else:
        dicta[word]=[]
        dicta[word].append(words[i+1])
```

```python
prob={}
for keys,values in dicta.items():
    prob[keys]=[]
    for ele in dicta[keys]:
        prob[keys].append((ele,float(dicta[keys].count(ele)/float(len(dicta[keys])))))
```

```python
prob['Joan'].sort()
prob['only'].sort()
prob['only']
```
