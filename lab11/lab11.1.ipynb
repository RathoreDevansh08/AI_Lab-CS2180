```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1=0.1
a2=0.01
a3=0.001
```

```python
def rec(al,y1,y2):  
    x=0
    X=[]
    T=[]
    X.append(x)
    T.append(0)
    for i in range(100):
        y=np.random.uniform(y1,y2)
        x=x+al*(y-x)
        X.append(x)
        T.append(i+1)
    return X,T
```

```python
x1,t1=rec(a1,-1,1)
plt.plot(t1,x1)
```

```python
def rec2(y1,y2):  
    x=0
    X=[]
    T=[]
    X.append(x)
    T.append(0)
    for i in range(100):
        y=np.random.uniform(y1,y2)
        al=1/(1+i)
        x=x+al*(y-x)
        X.append(x)
        T.append(i+1)
    return X,T
```

```python
x1,t1=rec2(0,1)
plt.plot(t1,x1)
```
