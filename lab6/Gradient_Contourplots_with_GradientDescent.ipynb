```
import numpy as np
import matplotlib.pyplot as plt
```

```
def J1(w1,w2):
  return w1*w1+w2*w2+5*w1-3*w2-2
```

```
def J11(w1,w2):
  return 2*w1+5
```

```
def J12(w1,w2):
  return 2*w2-3
```

```
def J2(w1,w2):
  return 10*w1*w1+w2*w2
```

```
def J21(w1,w2):
  return 20*w1
```

```
def J22(w1,w2):
  return 2*w2
```

```
m=10
n=20
```

```
X=np.linspace(-50,50,100)
Y=np.linspace(-50,50,100)
```

```
Z1,Z2=np.meshgrid(X,Y)
```

```
#Function 1:
```

```
Z=J1(Z1,Z2)
plt.contourf(Z1,Z2,Z)
```

```
U=J11(Z1,Z2)
V=J12(Z1,Z2)
plt.quiver(Z1,Z2,U,V)
```

```
def GD1(alpha=0.01):
  We=np.random.random((2,1))*30
  plt.contourf(Z1,Z2,Z)
  for i in range(1000):
    We[0][0]=We[0][0]-alpha*(J11(We[0][0],We[1][0]))
    We[1][0]=We[1][0]-alpha*(J12(We[0][0],We[1][0]))
    if(i%20==0):
      plt.plot(We[0][0],We[1][0],'ro')
  plt.plot(We[0][0],We[1][0],'go')
  plt.show()
  return We[0][0],We[1][0]
```

```
print("Gradient descent:")
W10,W11=GD1(0.01)
print("Final weights:")
print(W10,W11)
```
