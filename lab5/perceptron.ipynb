```
import numpy as np
import matplotlib.pyplot as plt
```

```
pro_k=0.5
```

```
X=np.ndarray((1000,3))
Y=np.ndarray((1000,1))
```

```
H=[]
W=[]

for i in range(1000):
  k=np.random.randint(10)
  if(k<pro_k*10):
    c=-1
    height=np.random.uniform(4.9,5.3)
    weight=np.random.uniform(30,45)
  else:
    c=1
    height=np.random.uniform(5.4,5.9)
    weight=np.random.uniform(50,65)
  H.append(height)
  W.append(weight)
  X[i][0]=height
  X[i][1]=weight
  X[i][2]=1
  Y[i][0]=c
```

```
plt.scatter(H,W)
```

```
def perceptron(alpha,it):
  We=np.random.random((3,1))
  We2=We[:]
  epoch=0
  for i in range(it):
    for j in range(1000):
      if((We[0][0]*X[j][0]+We[1][0]*X[j][1]+We[2][0])*(Y[j][0])<=0):
        We[0][0]=We[0][0]+alpha*((X[j][0]*Y[j][0]))
        We[1][0]=We[1][0]+alpha*((X[j][1]*Y[j][0]))
        We[2][0]=We[2][0]+alpha*((X[j][2]*Y[j][0]))
    if(i%40==0):
      t=np.linspace(4.5,6.0,1000)
      plt.plot(t,(-1)*(We[0][0]*t+We[2][0])/We[1][0],'r-')
  plt.scatter(H,W)
  plt.plot(t,(-1)*(We[0][0]*t+We[2][0])/We[1][0],'g-')
  plt.show()
  return We,We2
```

```
W1, W0=perceptron(0.0001,400)
print(W0.T)
print(W1.T)
```

```
def perceptron_withmychanges(alpha=(0.1),it=1000):
  #We=np.random.random((3,1))
  We=W0
  epoch=0
  for i in range(it):
    for j in range(1000):
      if((We[0][0]*X[j][0]+We[1][0]*X[j][1]+We[2][0])*(Y[j][0])<=0):
        We=We+alpha*(Y[j][0])*(X[j].T)/(i+1)                             #<---My proposed change
    if(i%40==0):
      t=np.linspace(4.5,6.0,1000)
      plt.plot(t,(-1)*(We[0][0]*t+We[2][0])/We[1][0],'r-')
  plt.scatter(H,W)
  plt.plot(t,(-1)*(We[0][0]*t+We[2][0])/We[1][0],'g-')
  plt.show()
  return We
```
